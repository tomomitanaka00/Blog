```python
#Bard's first attempt

villagers = [
    {"name": "Mark", "birthday": "October 1", "hobby": "nature", "species": "cat"},
    {"name": "Lucy", "birthday": "October 31", "hobby": "education", "species": "deer"},
    {"name": "Richard", "birthday": "January 1", "hobby": "play", "species": "cat"},
    {"name": "Matt", "birthday": "September 29", "hobby": "music", "species": "squirrel"},
    {"name": "Max", "birthday": "April 16", "hobby": "play", "species": "bear cub"},
]

# Create a new list to store the formatted data
formatted_villagers = []

# Loop through each villager in the original list
for villager in villagers:
    # Extract the name, birthday, hobby, and species
    name = villager["name"]
    birthday = villager["birthday"]
    hobby = villager["hobby"]
    species = villager["species"]

    # Combine the information into a single string
    formatted_data = f"{name} & {birthday} & {hobby} & {species}"

    # Append the formatted data to the new list
    formatted_villagers.append(formatted_data)

# Print the formatted list
for villager in formatted_villagers:
    print(villager)
```
